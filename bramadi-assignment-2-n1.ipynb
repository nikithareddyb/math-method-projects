```python
#importing all the required packages
%matplotlib inline
import random
import numpy as np
import seaborn as sns
import pandas as pd

## UNIFORM DISTRIBUTION ##

#generating random values ranging 0-25 using the uniform function of the random package
uni_random = [int(random.uniform(0,26))
       for i in range(1000)]
```

```python
#Generating a BAR GRAPH to check outcome
#performing a unique operation on the duplicated combination of random number and frequency counts
uniq = np.unique(uni_random, return_counts=True)

#Assigning the numpy array to the 2 dimensional array
values, frequencies = np.array(uniq)

#using seaborn package to create a bar graph for uniform distribution
#bar graph styling and labelling
sns.set_style('darkgrid')
uni_plot = sns.barplot(x=values, y=frequencies, color='purple')
uni_plot.set_title('Uniform Distribution of 1000 random numbers')
uni_plot.set(xlabel='Random integer of range 0-25',ylabel='Count of the integer')
```

```python
#using seaborn package to generate a histogram

#creating a dataframe from the list to be used in the histplot function
uni_random_df = pd.DataFrame(uni_random, columns=['uni_random_values'])

#Histogram graph styling and labelling
sns.set_style('whitegrid')
uni_hist_plot = sns.histplot(data=uni_random_df, x='uni_random_values',color='darkorange', bins=25, alpha=0.1, kde=True)
uni_hist_plot.set_title("Uniform Distribution of 1000 random number")
uni_hist_plot.set(xlabel='Random integer of range 0-25',ylabel= 'Frequency of the integer')
```
