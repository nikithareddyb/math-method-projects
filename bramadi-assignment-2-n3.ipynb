```python
## EXPONENTIAL DISTRIBUTION ##

#importing all the required packages
%matplotlib inline
import random
import numpy as np
import seaborn as sns
import pandas as pd

#Defining mean and standard deviation here:
mean=12.5

exp_random = [int(random.expovariate(1.0/mean))
       for i in range(1000)]

#create a new list with only values in range of 0-25
exp_random_filter=[] #creating an empty list
for value in exp_random:
    if value >= 0 and value <=25 :
        exp_random_filter.append(value)

#converting list into an numpy array
numpy_array_exp=np.array(exp_random_filter)

#creating a dataframe using pandas pd to generate the seaborn histogram
exp_df = pd.DataFrame(numpy_array_exp, columns =['values2'])
exp_plot = sns.histplot(data=exp_df, x='values2',color='navy', bins= 25, kde=True)
exp_plot.set_title('Exponential distribution curve of 1000 random numbers')
exp_plot.set(xlabel='Random number in range 0-25',ylabel='Frequency of occurence')
```

```python
#create a new list with only values in range of 0-100
exp_random_filter_1=[]
for value in exp_random:
    if value >= 0 and value <=100 :
        exp_random_filter_1.append(value)

#converting list into an numpy array
numpy_array_exp_1=np.array(exp_random_filter_1)

#creating a dataframe using pandas pd to generate the seaborn histogram
exp_df_1 = pd.DataFrame(numpy_array_exp_1, columns =['values3'])
exp_plot_1 = sns.histplot(data=exp_df_1, x='values3',color='navy', bins= 100, kde=True)
exp_plot_1.set_title('Exponential distribution curve of 1000 random numbers')
exp_plot_1.set(xlabel='Random number in range 0-100',ylabel='Frequency of occurence')
```
