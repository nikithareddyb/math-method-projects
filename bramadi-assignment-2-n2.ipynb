```python
#importing all the required packages
%matplotlib inline
import random
import numpy as np
import seaborn as sns
import pandas as pd

## NORMAL DISTRIBUTION ##

#Defining mean and standard deviation here:
mean=12.5
std_dev=5

#Generating 1000 random values using gauss function which produces normal distribution
normal_random = [int(random.gauss(mean,std_dev))
       for i in range(1000)]

#create a new list with only values in range of 0-25
normal_random_filter=[] #creating an empty list
for value in normal_random:
    if value >=0 and value <=25 :
        normal_random_filter.append(value)
```

```python
#converting the 0-25 range numbers list to numpy array
numpy_array = np.array(normal_random_filter)

#converting the numpy_array into a dataframe to be able to use in a histplot of seaborn
normal_random_df = pd.DataFrame(numpy_array, columns=['normal_random_values'])

#styling and labeling the graph
sns.set_style('darkgrid')
normal_hist_plot = sns.histplot(data=normal_random_df, x='normal_random_values',color='blue', kde=True, alpha= 0.2)
normal_hist_plot.set_title("Normal Distribution of 1000 random number")
normal_hist_plot.set(xlabel='Random integer in range 0-25',ylabel= 'Frequency of the integer')
```
